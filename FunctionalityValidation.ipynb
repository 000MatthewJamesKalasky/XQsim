# Functionality validation with (architecture + quantum) simuation
* This notebook aims to introduce how to validate the target QCP's functionality with a quantum simulation. Please check the code and result below.

## Import

```python
import os, sys
curr_dir = os.getcwd()
src_dir = os.path.join(curr_dir, "src")
sys.path.insert(0, src_dir)

from xqsim import xqsim
from visualization import draw_pqsim_res

from logical_simulation import run_ppr_with_qiskit
```

## Setup

```python
config = "example_cmos_d5"
qbin = "pprIIZZZ_n5"
num_shots = 2048
```

## Run XQsim

```python
framework = xqsim()
framework.setup(config = config, 
                qbin = qbin,
                num_shots = num_shots, 
                dump_synth = False,
                regen_synth = False,
                dump_est = False, 
                regen_est = False, 
                dump_sim = False, 
                regen_sim = False, 
                debug = False
                )
_, pqsim_res = framework.run()
```

## Run logical-qubit level quantum simulation

```python
lqsim_res = run_ppr_with_qiskit(ppr="IIZZZ", num_shots=num_shots)
```

## Visualize & Compare quantum state distribution

```python
draw_pqsim_res(pqsim_res, lqsim_res)
```
